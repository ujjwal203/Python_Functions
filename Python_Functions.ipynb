1) **What is the difference between a function and a method in Python?**
- A function in Python is a block of reusable code that performs a specific task.
- It is defined using the def keyword or lambda.
  - def greet(name):
  -    return f"Hello, {name}!"
  - print(greet("Ujjwal"))   # Calling the function directly

- A method is a function that belongs to an object.
- Methods are defined inside a class and are called on instances (objects) of that class.
  - class Person:
  - def __init__(self, name):
  -    self.name = name
  - def greet(self): #method
  -   return f"Hello, {self.name}!"

  - p = Person("Ujjwal") # Create an object
  - print(p.greet())   # Calling a method on an object

4) **What is the purpose of the `return` statement in a Python function?**
- The return statement is used inside a function to:
- Send a value back to the caller (the place where the function was invoked).
- End the function’s execution immediately (no code after return is run).

5) **What are iterators in Python and how do they differ from iterables?**
- An iterable is any Python object that can return its members one at a time.
- Examples: lists, tuples, strings, sets, dicts, files, ranges.
- Whereas;
- An iterator is the object that actually produces values one at a time from an iterable.

6) **Explain the concept of generators in Python and how they are defined**
- A generator in Python is a special type of iterator that generates values on the fly instead of storing them all in memory.
- Generators are defined like a normal function, but with yield.
- Example code :
- def simple_gen():
  -   yield 1
  -   yield 2
  -   yield 3

- gen = simple_gen()
- print(next(gen))  # 1
- print(next(gen))  # 2
- print(next(gen))  # 3

7) **What are the advantages of using generators over regular functions?**
- A regular function (def) computes all results and returns them at once (often as a list).
A generator function (def with yield) produces results one at a time, lazily, only when requested.

- Advantage:
- 1. Memory Efficiency : Regular functions build the entire result in memory.
- 2. Performance (Faster for Large Data) : Since values are generated on demand, you don’t wait for the entire sequence.
- 3. Infinite Sequences Possible : Generators can model infinite streams (e.g., Fibonacci, sensor data).
- 4. Pipeline Processing : Generators can be chained to process data step by step

8) **What is a lambda function in Python and when is it typically used?**
- A lambda function in Python is an anonymous (nameless) function created using the lambda keyword. It can have any number of arguments but only one expression.
The expression is evaluated and returned automatically.

- Lambda functions are used for short, throwaway operations — when you don’t want to formally define a function with def.

9) **Explain the purpose and usage of the `map()` function in Python.**
- The map() function is used to apply a function to every item of an iterable (like a list, tuple, or string) and return a new map object (iterator) with the results.
It’s mainly for transformation or changing each element without writing explicit loops.

10) **What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?**
- map (); Applies a function to every element in an iterable and returns a new iterable (map object)
- reduce(); Applies a function (predicate returning True/False) to each element and keeps only those that return True.
- filter(); Applies a function cumulatively to the elements of an iterable, reducing it to a single value.
